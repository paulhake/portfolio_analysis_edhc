```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import edhec_risk_ph as erk
%load_ext autoreload
%autoreload 2
```

```python
def discount(t, r):
    """Compute discount factor for time t and rate r."""
    return 1 / (1 + r) ** t
```

```python
discount(1/4,0.04)
```

```python
def pv(cash_flows, r):
    """Compute present value of a series of cash flows given rate r."""
    periods = cash_flows.index
    discounts = discount(periods, r)
    return np.sum(cash_flows * discounts)
```

```python
cash_flows = pd.Series(data=[100, 100, 100, 100, 100],index=[1,2,3,4,60])
pv(cash_flows, 0.05)
```

```python
cash_flows
```

```python
def funding_ratio(liabilities, assets, r):
    """Compute funding ratio given liabilities and assets present values."""
    pv_liabs = pv(liabilities, r)
    #pv_assets = pv(assets, r)
    return assets / pv_liabs
```

```python
funding_ratio(cash_flows, 300, 0.15)
```

```python
import ipywidgets as widgets

def show_funding_ratio(assets, r):
    fr = funding_ratio(cash_flows, assets, r)
    print(f"Funding ratio at rate {r:.2%} is {fr:.2%}")

widgets.interact(show_funding_ratio, 
                r=widgets.FloatSlider(value=0.05, 
                                        min=0.0, 
                                        max=0.15, 
                                        step=0.01),
                assets=widgets.IntSlider(min=1,
                                        max=500,
                                        step=1,
                                        value=300)
                )
```
