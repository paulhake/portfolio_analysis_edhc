```python
import numpy as np
import pandas as pd

def gbm0(n_years=10, n_scenarios=1000, mu=0.07, sigma=0.15, steps_per_year=12, s0=100):
    

    # Time parameters
    dt = 1/steps_per_year  # 1 year time step
    n_steps = int(n_years*steps_per_year)

    xi = np.random.normal(0, 1, (n_steps, n_scenarios))
    rets = mu * dt + sigma * np.sqrt(dt) * xi
    rets = pd.DataFrame(rets)
    prices = s0 * (1 + rets).cumprod()
    return prices
```

```python
def gbm(n_years=10, n_scenarios=1000, mu=0.07, sigma=0.15, steps_per_year=12, s0=100):
    

    # Time parameters
    dt = 1/steps_per_year  # 1 year time step
    n_steps = int(n_years*steps_per_year)

    rets_plus_one = np.random.normal(loc=1+mu*dt, scale=sigma * np.sqrt(dt), size=(n_steps+1, n_scenarios))
    rets_plus_one[0] = 1  # set first row to 1s for initial price level
    prices = pd.DataFrame(rets_plus_one).cumprod() * s0
    
    return prices
```
